```python
#Import dependencies
import pandas as pd
import os
```

```python
#Filepath to teams for df
teams_filepath = os.path.join('..', 'Resources', 'teams.csv')

messy_teams_df = pd.read_csv(teams_filepath)

# Header Cleaner
messy_teams_df.columns = messy_teams_df.columns.str.strip()\
    .str.lower()\
    .str.replace(' ', '_')\
    .str.replace('.', '')\
    .str.replace('(', '')\
    .str.replace(')', '')
    # ... add .str.replace('','') for other characters
messy_teams_df.head()
```

```python
#Drop unneeded columns 
clean_teams_df = messy_teams_df.drop(['league_id', 'min_year', 'max_year', 'yearfounded', 'arenacapacity', 'owner', 'generalmanager', 'headcoach', 'dleagueaffiliation'], axis = 1)

clean_teams_df.to_csv('../outputs/teams_cleaned.csv', index = False)
```

```python
#Filepath to games
games_filepath = os.path.join('..', 'Resources', 'games.csv')
messy_games_df = pd.read_csv(games_filepath)

# Header Cleaner
messy_games_df.columns = messy_games_df.columns.str.strip()\
    .str.lower()\
    .str.replace(' ', '_')\
    .str.replace('.', '')\
    .str.replace('(', '')\
    .str.replace(')', '')
    # ... add .str.replace('','') for other characters
messy_games_df.head()
```

```python
messy_games_df.game_date_est = pd.to_datetime(messy_games_df.game_date_est)
messy_games_df
```

```python
#Access games only in 2018
games_2018_df = messy_games_df.loc[messy_games_df.season == 2018]
```

```python
#Rearrange columns
cleaned_games_2018 = games_2018_df[['game_id', 'game_date_est', 'home_team_id', 'visitor_team_id', 'season', 'pts_home', 'fg_pct_home', 'ft_pct_home', 'fg3_pct_home', 'ast_home', 'reb_home', 'pts_away', 'fg_pct_away', 'ft_pct_away', 'fg3_pct_away', 'ast_away', 'reb_away', 'home_team_wins']]
```

```python
#Export to csv
cleaned_games_2018.to_csv('../outputs/games_cleaned.csv', index = False)
```
