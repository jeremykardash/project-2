```python
import pandas as pd
import os
```

```python
filepath = os.path.join("..",'Resources', 'nba_draft_combine_all_years.csv')

combine = pd.read_csv(filepath)

# Header Cleaner
messy_teams_df.columns = messy_teams_df.columns.str.strip()\
    .str.lower()\
    .str.replace(' ', '_')\
    .str.replace('.', '')\
    .str.replace('(', '')\
    .str.replace(')', '')
    # ... add .str.replace('','') for other characters
combine.head()
```

```python
file = os.path.join("..",'Resources', 'players.csv')

players = pd.read_csv(file)
players
```

```python
players = players.loc[players['SEASON'] == 2018]
```

```python
merged = pd.merge(combine, players, left_on="Player", right_on="PLAYER_NAME")
```

```python
list(merged)
```

```python
clean_df = merged[['PLAYER_ID','Player','Year','TEAM_ID',
 'Draft pick',
 'Height (No Shoes)',
 'Height (With Shoes)',
 'Wingspan',
 'Standing reach',
 'Vertical (Max)',
 'Vertical (Max Reach)',
 'Vertical (No Step)',
 'Vertical (No Step Reach)',
 'Weight',
 'Body Fat',
 'Hand (Length)',
 'Hand (Width)',
 'Bench',
 'Agility',
 'Sprint',
 'SEASON']]
```

```python
clean_df
```

```python
clean_df.columns = clean_df.columns.str.strip()\
    .str.lower()\
    .str.replace(' ', '_')\
    .str.replace('.', '')\
    .str.replace('(', '')\
    .str.replace(')', '')
    # ... add .str.replace('','') for other characters
clean_df.head()
```
