```python
import pandas as pd
import os
```

```python
teams_filepath = os.path.join('..', 'Resources', 'player_stats.csv')

data = pd.read_csv(teams_filepath)

# Header Cleaner
data.columns = data.columns.str.strip()\
    .str.lower()\
    .str.replace(' ', '_')\
    .str.replace('.', '')\
    .str.replace('(', '')\
    .str.replace(')', '')
    # ... add .str.replace('','') for other characters
data.head()
```

```python
# reading csv file from url  

# dropping null value columns to avoid errors 
data.dropna(inplace = True) 
  
# new data frame with split value columns 
new = data["player"].str.split("\\", n = 1, expand = True)  
  
# making separate first name column from new data frame 
data["Name"]= new[0] 
  
# making separate last name column from new data frame 
  
# Dropping old Name columns 
  
# df display 
data
```

```python
clean_stats = data[['Name',
      'rk',
 'pos',
 'age',
 'tm',
 'g',
 'gs',
 'mp',
 'fg',
 'fga',
 'fg%',
 '3p',
 '3pa',
 '3p%',
 '2p',
 '2pa',
 '2p%',
 'efg%',
 'ft',
 'fta',
 'ft%',
 'orb',
 'drb',
 'trb',
 'ast',
 'stl',
 'blk',
 'tov',
 'pf',
 'pts']]

```

```python
clean_stats = clean_stats.drop_duplicates(subset=['Name'], keep='first')
clean_stats
```

```python
file = os.path.join("..",'Resources', 'players.csv')

players = pd.read_csv(file)
players
```

```python
players = players.loc[players['SEASON'] == 2018]
players
```

```python
merged = pd.merge(clean_stats, players, left_on="Name", right_on="PLAYER_NAME")
```

```python
merged
```

```python
cleaned_df = merged.drop_duplicates(subset=['Name'], keep='first')
```

```python
list(cleaned_df)
```

```python
stats = cleaned_df[['PLAYER_ID',
 'TEAM_ID', 'Name',
 'rk',
 'pos',
 'age',
 'tm',
 'g',
 'gs',
 'mp',
'pts',
 'fg',
 'fga',
 'fg%',
 '3p',
 '3pa',
 '3p%',
 '2p',
 '2pa',
 '2p%',
 'efg%',
 'ft',
 'fta',
 'ft%',
 'orb',
 'drb',
 'trb',
 'ast',
 'stl',
 'blk',
 'tov',
 'pf',
 'SEASON']]
```
