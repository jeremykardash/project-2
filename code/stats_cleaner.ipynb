```python
import pandas as pd
import os
```

```python
teams_filepath = os.path.join('..', 'Validated_Data', 'player_stats_val.csv')

data = pd.read_csv(teams_filepath)

data.head()
```

```python
# reading csv file from url  

# dropping null value columns to avoid errors 
data.dropna(inplace = True) 
  
# new data frame with split value columns 
new = data["player"].str.split("\\", n = 1, expand = True)  
  
# making separate first name column from new data frame 
data["name"]= new[0] 
  
# df display 
data
```

```python
#rearrange columns
clean_stats = data[['name',
'rk',
 'pos',
 'age',
 'tm',
 'g',
 'gs',
 'mp',
 'fg',
 'fga',
 'fg%',
 '3p',
 '3pa',
 '3p%',
 '2p',
 '2pa',
 '2p%',
 'efg%',
 'ft',
 'fta',
 'ft%',
 'orb',
 'drb',
 'trb',
 'ast',
 'stl',
 'blk',
 'tov',
 'pf',
 'pts']]

```

```python
#drop duplicates due to merge
clean_stats = clean_stats.drop_duplicates(subset=['name'], keep='first')
clean_stats['name'] = clean_stats['name'].str.replace('.','')
clean_stats
```

```python
#filepath and df to players
file = os.path.join("..",'Validated_Data', 'players_val.csv')

players = pd.read_csv(file)
players
```

```python
#reduce players for season and edit their names for consistency
players = players.loc[players['season'] == 2018]

players["player_name"] = players["player_name"].str.replace('.', '')
```

```python
#merge the data
merged = pd.merge(clean_stats, players, left_on="name", right_on="player_name")
```

```python
merged
```

```python
cleaned_df = merged.drop_duplicates(subset=['name'], keep='first')
```

```python
list(cleaned_df)
```

```python
stats = cleaned_df[['player_id',
 'team_id', 'name',
 'pos',
 'age',
 'tm',
 'g',
 'gs',
 'mp',
 'fg',
 'fga',
 'fg%',
 '3p',
 '3pa',
 '3p%',
 '2p',
 '2pa',
 '2p%',
 'efg%',
 'ft',
 'fta',
 'ft%',
'pts',
 'orb',
 'drb',
 'trb',
 'ast',
 'stl',
 'blk',
 'tov',
 'pf',
 'season']]
```

```python
#rename columns
stats = stats.rename(columns={
    "name": "player_name",
    "fg%":"fg_percent",
    "3p":"ThreeP_m",
    "3pa":"ThreeP_a",
    "3p%":"ThreeP_percent",
    "2pa":"TwoP_a",
    "2p":"TwoP_m",
    "2p%":"TwoP_percent",
    "efg%":"efg_percent",
    "ft%":"ft_percent"
})
```
